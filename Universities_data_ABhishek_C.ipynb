```python
import string
import sqlite3
import re
import time
from selenium import webdriver

driver = webdriver.Chrome(r'C:\Users\ADMIN\anaconda3\Lib\site-packages\selenium\chromedriver.exe')
URL_pattern_str = "https://cwur.org/2022-23.php"
#page_URL=URL_pattern_str.replace("$NUM$",str())
page_URL = URL_pattern_str
print("Start with",'https://cwur.org/2022-23.php')
driver.get(page_URL)
page_content = driver.page_source
```

```python
from selenium.webdriver.common.by import By
table = driver.find_element(By.XPATH,"//*[@id='cwurTable']")
rows = table.find_elements(By.TAG_NAME, "tr")
conn = sqlite3.connect('Universities.db')
c = conn.cursor()
c.execute("CREATE TABLE IF NOT EXISTS Uni(World_Rank VARCHAR(100), Institution VARCHAR(20), Location VARCHAR(50), National_Rank VARCHAR(10), Education_Rank VARCHAR(50), Employability_Rank VARCHAR(300),Faculty_Rank VARCHAR(300),Research_Rank VARCHAR(100),Score VARCHAR(100))")
for row in rows[1:]:
    cells = row.find_elements(By.TAG_NAME, "td")
    World_Rank = cells[0].text
    Institution = cells[1].text
    Location = cells[2].text
    National_Rank = cells[3].text
    Education_Rank = cells[4].text
    Employability_Rank = cells[5].text
    Faculty_Rank = cells[6].text
    Research_Rank = cells[7].text
    Score = cells[8].text
    insert_query = '''INSERT INTO Uni(World_Rank, Institution, Location, National_Rank, Education_Rank, Employability_Rank, Faculty_Rank,Research_Rank,Score)
                 VALUES (?, ?, ?, ?, ?,?,?,?,?)''', (World_Rank, Institution, Location, National_Rank, Education_Rank, Employability_Rank, Faculty_Rank,Research_Rank,Score)
    c.execute(*insert_query)

    
```

```python
import pandas as pd
c.execute('SELECT*FROM Uni')
df = pd.DataFrame(c.fetchall(), columns = ['World_Rank', 'Institution', 'Location', 'National_Rank', 'Education_Rank', 'Employability_Rank', 'Faculty_Rank','Research_Rank','Score'])
print (df)
```
